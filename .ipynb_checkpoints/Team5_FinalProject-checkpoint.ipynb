## Data Splitting

```python
# One-hot encode the categorical variables
food_encoded_data = pd.get_dummies(food, columns=['Day of the Week', 'Cuisine Type', 'Restaurant Name'])

print(food_encoded_data.columns)
```

```python
# threshold for binary classification (4-5 positive, negative is ratings less than 4)
threshold = 4

# Create a new binary target variable based on the threshold
food_encoded_data['Binary_Rating'] = (food_encoded_data['Rating'] >= threshold).astype(int)
```

```python
# defining features(X) and target (y) variables
X = food_encoded_data.drop(['Rating', 'Binary_Rating'], axis=1)
y = food_encoded_data['Binary_Rating']

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# calculate proportions
total_samples = len(X)
train_samples = len(X_train)
test_samples = len(X_test)

train_proportion = round((train_samples / total_samples) * 100)
test_proportion = round((test_samples / total_samples) * 100)

# print proportions
print(f"Training Proportion: {train_proportion}%")
print(f"Test Proportion: {test_proportion}%")
```

**The Rating is created into a binary rating with values of 5 positive and anything under negative. Then the categorical variables are one-hot encoded and then the data is split into a training and test data set. The test data set makes up 30% of the data while the training data set is 70% of the data set. The split data is then checked to ensure that the splitting was done correctly.**

## Modeling/Evaluation

### Random Forest

```python
# Model 1: Random Forest Classifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)
rf_predictions = random_forest_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)

print("Random Forest Model Accuracy:", rf_accuracy)
print("Classification Report for Random Forest Model:")
print(classification_report(y_test, rf_predictions))
```

### Logistic Regression

```python
# Model 2: Logistic Regression
logistic_regression_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_regression_model.fit(X_train, y_train)
lr_predictions = logistic_regression_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)

print("Logistic Regression Model Accuracy:", lr_accuracy)
print("Classification Report for Logistic Regression Model:")
print(classification_report(y_test, lr_predictions))
```

### Support Vector Machine

```python
# Model 3: Support Vector Machine (SVM)
svm_model = SVC(kernel='linear', random_state=42, probability=True)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)

print("Support Vector Machine (SVM) Model Accuracy:", svm_accuracy)
print("Classification Report for SVM Model:")
print(classification_report(y_test, svm_predictions))
```

**The goals of these models were to perform binary classification on the dataset. The goals of the models is to predict whether a given food order is satisfactory (labeled as 1) or not (labeled as 0). Orders that received a rating of 4 were considered satisfactory, and under that is unsatisfactory. The Random Forest Classifier achieved an accuracy of approximately 81%. It demonstrated strong precision and recall for class 1 (satisfactory orders), but lower values for class 0 (unsatisfactory orders). The Logistic Regression model attained an accuracy of roughly 83%. It had high precision, recall, and F1-score for class 1 but struggled with class 0, which might be attributed to the class imbalance in the dataset. The Support Vector Machine (SVM) also delivered an accuracy of around 83%. It displayed improved precision and recall for class 0 compared to the other models but still excelled at classifying class 1. The dataset appeared to be imbalanced, with a disproportionate number of class 1 samples (satisfactory orders) compared to class 0 (unsatisfactory orders). As a result, the models tend to perform well on the majority class but not as effectively on the minority class. SVM so far would be our best model to predict satisfactory and unsatisfactory classes.**

### AUC-ROC curves

```python
model_probs = {
    'Random Forest': random_forest_model.predict_proba(X_test)[:, 1],
    'Logistic Regression': logistic_regression_model.predict_proba(X_test)[:, 1],
    'SVM': svm_model.decision_function(X_test)
}

plt.figure(figsize=(10, 6))

# Plot ROC curves and calculate AUC for each model
for model_name, y_pred_probs in model_probs.items():
    fpr, tpr, _ = roc_curve(y_test, y_pred_probs)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

**The output displays Receiver Operating Characteristic (ROC) curves for three different machine learning models: Random Forest, Logistic Regression, and Support Vector Machine (SVM).The curves are used to evaluate performance of binary classification models. A higher AUC indicates a better model performance. In this case, the Random Forest and Logistic Regression models have similar AUC values of approximately 0.52, indicating that they have similar discriminative power. The SVM model has a lower AUC of 0.45, suggesting it might not perform as well in distinguishing between the two classes. It is important to recall that this dataset is imbalanced.**

### Cummulative Gains Plot

```python
# Create a DummyClassifier that predicts the majority class
baseline_model = DummyClassifier(strategy='most_frequent')

# Fit the baseline model to the training data
baseline_model.fit(X_train, y_train)

# Predict majority class for all instances in the test data
baseline_predictions = baseline_model.predict(X_test)

# Evaluate baseline model
baseline_accuracy = accuracy_score(y_test, baseline_predictions)
print("Baseline Model Accuracy:", baseline_accuracy)

# Print classification report
print("Classification Report for Baseline Model:")
print(classification_report(y_test, baseline_predictions))
```

```python
# Sort the models by descending order of predicted probabilities
models = [random_forest_model, logistic_regression_model, svm_model, baseline_model]

# Create an array to store the cumulative gains for each model
cumulative_gains = []

# Calculate the cumulative gains for each model
for model in models:
    y_pred_probs = model.predict_proba(X_test)[:, 1]
    sorted_indices = np.argsort(y_pred_probs)[::-1]
    y_sorted = y_test.iloc[sorted_indices]
    cumulative_gain = np.cumsum(y_sorted) / np.sum(y_test)
    cumulative_gains.append(cumulative_gain)

# Plot the cumulative gains curves
plt.figure(figsize=(10, 6))
for i, model_name in enumerate(["Random Forest", "Logistic Regression", "SVM", "Baseline"]):
    plt.plot(np.linspace(0, 1, len(cumulative_gains[i])), cumulative_gains[i], label=model_name)

plt.xlabel("Proportion of Test Instances")
plt.ylabel("Cumulative Gains")
plt.title("Cumulative Gains Chart for Different Models")
plt.legend()
plt.grid(True)
plt.show()
```

**The basline model was created as a reference to assess the performance of the other models. The baseline model does not have an predictive power. The cumulative gains curves for the three models created have curves similar to the baseline indicating that these models are not providing significant improvement over random choice in predicting positive instances.**

**In this food delivery analysis, we explored different machine learning models to predict whether food orders are satisfactory or not based on various factors. We considered Random Forest, Logistic Regression, and Support Vector Machine (SVM) models.**

**After evaluating these models, we found that SVM performed the best. It showed the highest accuracy and proved to be a reliable choice for classifying food orders. SVM's predictions were more precise in distinguishing between satisfactory and unsatisfactory orders compared to the other models.**

**If you're looking to use a model to help predict customer happiness, we would recommend the Support Vector Machine (SVM) approach. It's like having a trustworthy guide to make informed decisions and ensure customer satisfaction. SVM provides a balance between precision and recall, making it a valuable tool for food delivery businesses to enhance their service quality and customer experience.**

### K-Means Clustering

```python

# Selecting all one-hot encoded features
selected_features = food_encoded_data.drop(['Cost of the Order', 'Rating', 'Food Preparation Time', 'Delivery Time'], axis=1)

# Determine the optimal k using the elbow method
wcss = []  

# Try different values of k and calculate the WCSS for each k
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(selected_features)
    wcss.append(kmeans.inertia_)

# Plot the within-cluster sum of squares for different values of k
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
# Define the number of clusters (k)
k = 3

# Create a K-Means model with k clusters
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model to your entire dataset (X_train and X_test combined)
kmeans.fit(pd.concat([X_train, X_test]))

# Predict the cluster labels for the test data
kmeans_predictions = kmeans.predict(X_test)

# Add the cluster labels to your test data
X_test['Cluster'] = kmeans_predictions

# Now you can plot the clusters as previously shown
for i in range(k):
    plt.scatter(X_test[X_test['Cluster'] == i]['Cost of the Order'], X_test[X_test['Cluster'] == i]['Food Preparation Time'], label=f'Cluster {i}')

plt.xlabel('Cost of the Order')
plt.ylabel('Food Preparation Time')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```
